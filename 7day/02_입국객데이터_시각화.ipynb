```python
### 우리나라에 방문하는 외국인 관광객이 가장 많은 국가 top5 조회하기
# - 각 국적별로 집계(그룹)한 후 -> 집계된 전체관광객 비율의 값을 평균하여 
# - 평균이 가장 높은 순으로 정렬 후 top5 데이터만 추출
temp = df.pivot_table(
    # 집계할 기준 [컬럼명] 정의
    index="국적",
    #하나인 경우 " ", 여러개인 경우 [] 리스트로
    values="전체관광객비율(%)",
    # 그룹을 대표하는 데이터 값은 : 평균을 사용 
    aggfunc="mean"
)
temp

### 정렬하기
# - 오름차순 : 낮은 값부터 ~ 높은 값 순으로 정렬(ascending, ASC)
# - 내림차순 : 높은 값부터 ~ 낮은 값 순으로 정렬(descending,DESC)
temp = temp.sort_values("전체관광객비율(%)",ascending = False)
temp

### Top 5 
temp = temp.head(5)
temp
```

```python
### 각 국적에 대한  중국 ,일본,대만,미국,홍콩 제목 5개 
### 행 인덱스 값(보이는 인덱스 값)을 컬럼으로 사용하기
# - 인덱스를 컬럼으로 변경하기
### 처리한 데이터프레임의 결과를 변수에 담지 않고, 직접 메모리에 저장시키는 방법
# - inplace = True를 파라메터로 정의
# 
temp.reset_index(inplace = True)
```

### 히트맵 그래프 그리기

```python
"""
선그래프 : x, y 데이터 사용
히트맵   : x, y, z(시점 데이터) 데이터 사용
 - 히트맵은 좀 더 자세한 시점의 값들을 색상으로 확인 가능
"""
```

```python
### 원본 데이터 
# - copy() : df의 메모리를 복제하여 새로운 메모리를 만들고, 해당 주소를 받음 
#          : df와 분리됨(df의 메모리 주소를 바라보지 않음)

df2 = df.copy()
df.head(1)
```

```python
### 히트맵에 사용할 데이터
# - x축 : "기준년월" 데이터에서 "월" 데이터 사용
# - y축 : "기준년월" 데이터에서 "년도" 데이터 사용
# - 시점 데이터 : 관광 데이터 ==> '대표값을 써야함' - 평균 값
```

```python
### 히트맵 시각화시 -> 피벗 테이블로 집계하면 편함
# - 년도 및 월별로 관광 데이터의 평균 집계 : 피벗 테이블 
```

```python
### x축 및 y축에 들어갈 년도 및 월 데이터 생성하기 
# y축 년도 데이터는 y_list 변수에
# x축 년도 데이터는 x_list 변수에
df2["기준년월"][0][0:4], df2["기준년월"][0][5:7]
df2["기준년월"][1][0:4], df2["기준년월"][1][5:7]

### y축에 사용할 데이터를 담을 리스트 변수
y_list = []
### x축에 사용할 데이터를 담을 리스트 변수
x_list = []

### df2의 기준년도 컬럼 데이터에서 년도 및 월 데이터를 각각 추출하여
#   - 각각의 리스트 변수에 추가
for idx in range(len(df2)) : 
    y = df2["기준년월"][idx][0:4]
    x = df2["기준년월"][idx][5:7] 
    y_list.append(y)
    x_list.append(x)
    
# print(y_list)
# print()
# print(x_list)
```

```python
### df2 데이터프레임에 "년도" 및 "월" 컬럼 생성

df2["년도"] = y_list
df2["월"] = x_list

df2.head(1)
# df2.columns
```

```python
### 기준년월 컬럼의 데이터를 이용해서
# - 년도 컬럼 생성 후 년도 데이터 추가
# - 월 컬럼 생성 후 월 데이터 추가
df2["년도"] = df2["기준년월"].str.slice(0,4)
df2["월"] = df2["기준년월"].str.slice(5,7)
df2.head(2)
```

```python
### 샘플링
c = df2["국적"] == "중국"
df_fillter = df2[c]
df_fillter.head(1)
```

```python
### 피벗테이블로 집계하기 : 히트맵에 표시할 데이터 형태로 나타내기에 좋음
df_pivot = df_fillter.pivot_table(
    # - y축은 인덱스
    index="년도",
    # - x축은 컬럼
    columns="월",
    # - 집계할 데이터(시점 데이터)
    values="관광",
    # - 집계 방법(집계 함수라고 함)
    aggfunc="mean"
)
df_pivot
```

```python
### 시각화 라이브러리
# - 파이썬에서 사용되는 기본 시각화 라이브러리
import matplotlib.pyplot as plt

# - 히트맨 라이브러리
import seaborn as sns

# - 한글처리
plt.rc("font", family = "Malgun Gothic")

# - 마이너스 기호 깨짐 처리
plt.rcParams["axes.unicode_minus"] = False
```

```python
### 히트맵 그리기
# - 그래프 사이즈 
plt.figure(figsize=(16,10))

# - 그래프 제목
plt.title("년도 및 월별 [중국]관광객 추이 현황 히트맵 시각화")

# - 히트맵 그래프 그리기
# - annot : 시점 데이터 값을 보여줄지(True) 말지(False)
# - fmt : 실수형 데이터의 소숫잠 자릿수 결정(.0f는 소숫점 밑으로 사용하지 않음)
# - cmap : 색상맵이라고 칭함
sns.heatmap(df_pivot, annot=True, fmt=".0f", cmap = "rocket_r")

# - 그래프 보여주고, 메모리 비워
plt.show()
```
