#### 데이터 통합하기

```python
"""
<데이터 통합 방법>
 - 다른 파일 또는 데이터프레임의 데이터를 열(column) 또는 행(Row) 단위로
   추가하는 방식을 "통합"이라고 합니다.
 - 데이터 통합은 열단위통합과 행단위통합 2가지 통합을 수행

 (열단위 통합 // 세로)
  - 외부 파일 또는 데이터프레임에 있는 특정 열(column)의 데이터를 원본으로 가지고 올 때 사용
  - 보통 "메타정의서"라는 파일의 데이터를 원본으로 가지고 올 때 주로 사용되는 통합 방식
  - 원본의 행의 갯수와 통합하려는 외부 파일의 컬럼의 데이터(행)의 갯수는 같아야 함
    또는 같지 않더라도, 특정 값이 같은 조건의 관계에 의해서 가져오는 경우에도 사용됩니다.
  - 열단위 통합에 주료 사용되는 함수 : merge()
    
 (행단위 통합 // 가로)
  - 수집한 데이터의 양이 너무 많은 경우, 여러개의 파일로 분리해서 제공하는 경우가 많다.
  - 이런 경우 분리된 모든 파일을 하나로 통합 할 때 사용하는 방식으로, 
  - 원본과 외부 파일의 컬럼의 성격과 컬럼의 갯수가 일치해야 합니다.
  - 행단위 통합에 주로 사용되는 함수 : concat([원본, 추가])
"""
```

```python
### 국적코드에 대한 국적명이 정의된 외부 파일 가져오기
file_path = "./files/sample_codemaster.xlsx"
code_master = pd.read_excel(file_path)
code_master
```

```python
### 결측데이터 확인
# - 데이터 프레임정보를 확인하려면 결측데이터 1차적으로 확인 가능
code_master.info()
```

```python
"""
 - left : 왼쪽 메모리 영역으로, 보통 원본 데이터프레임이 위치함
 - right : 오른쪽 메모리 영역으로, 보통 외부 데이터프레임이 위치함
           (가져올 데이터가 있는 곳)
 - left_on과 right_on : 왼쪽 데이터와 오른쪽 데이터를 통합 시 매핑할 컬럼명을 작성
                        (보통 코드 데이터로 되어 있는 경우가 많으며,
                         컬럼의 성격이 같은 경우 사용 가능)
 - how : 통합방식 
         -> inner : 왼쪽 및 오른쪽의 매핑 컬럼의 값이 같은 데이터의 행들만 추출
                  : 오른쪽 데이터를 -> 왼쪽으로 통합
         -> left  : 왼쪽 전체 행을 추출(단, 매핑된 결과가 없으면 NaN으로 처리)
                  : 오른쪽 데이터를 -> 왼쪽으로 통합 
         -> right : 오른쪽 전체 행을 추출(단, 매핑된 결과가 없으면 NaN으로 처리)
                  : 왼쪽 데이터를 -> 오른쪽으로 통합
"""

pd.merge(left=sample_1,
        right=code_master,
        how="inner",
        left_on="국적코드",
        right_on="국적코드")
```

```python
pd.merge(left=sample_1,
        right=code_master,
        how="left",
        left_on="국적코드",
        right_on="국적코드")
```
